```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False,ax=ax)

plt.xlabel("Column")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np
books["missing_description"]= np.where(books["description"], 1,0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman") # spearmen when it is non continous value

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,6))

heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"} )
heatmap.set_title("Correlation Matrix Heatmap")

plt.show()
```

```python
books_missing_25_word = books_missing[books_missing["words_in_description"]>=25]
```

```python
books_missing_25_word
```

```python
books_missing_25_word["title_and_subtitle"] = (
    np.where(books_missing_25_word["subtitle"].isna(), books_missing_25_word["title"],
             books_missing_25_word[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
books_missing_25_word
```

```python
books_missing_25_word["tagged_description"]= books_missing_25_word[["isbn13", "description"]].astype(str).agg(": ".join, axis=1)
```

```python
(
    books_missing_25_word
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
