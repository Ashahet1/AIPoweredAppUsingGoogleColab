```python
catergory_mapping ={'Fiction': "Fiction",
                    'Juvenile Fiction': "Children's Fiction",
                    'Biogragphy and Autobiography': "NonFiction",
                    'History': "NonFiction",
                    'Literary': "NonFiction",
                    'Philosophy': "NonFiction",
                    'Religion': "NonFiction",
                    'Comics & Graphic Novel': "Fiction",
                    'Drama': "Fiction",
                    'Juvenile NonFiction': "Children's NonFiction",
                    'Science': "NonFiction",
                    'Poetry': "Fiction"
}
```

```python
books["simple_categories"] = books["categories"].map(catergory_mapping)
```

```python
from transformers import pipeline
fiction_categories = ["Fiction", "NonFiction"]
pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli",
                device=-1)
```

```python
sequences=books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
import numpy as np

max_index = np.argmax(pipe(sequences, fiction_categories)["scores"])
max_label = pipe(sequences, fiction_categories)["labels"][max_index]
max_label = pipe(sequences, fiction_categories)["labels"][max_index]
max_label
```

```python
def generate_predictions(sequences, categories):
    predictions = pipe(sequences, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_categories = []
predict_categories = []

for i in tqdm(range(0,300)):
    sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
    predict_categories += [generate_predictions(sequence, fiction_categories)]
    actual_categories += ["Fiction"]
```

```python
for i in tqdm(range(0,300)):
    sequence = books.loc[books["simple_categories"] == "NonFiction", "description"].reset_index(drop=True)[i]
    predict_categories += [generate_predictions(sequence, fiction_categories)]
    actual_categories += ["NonFiction"]
```

```python
prediction_df = pd.DataFrame({"actual_categories": actual_categories, "predict_categories": predict_categories})
```

```python
prediction_df
```

```python
prediction_df["correct_categories"] = (
    np.where(prediction_df["actual_categories"] == prediction_df["predict_categories"], 1,0)
)
```

```python
isbns = []
prediction_cat = []

missing_cat = books.loc[books["simple_categories"].isna(), ["isbn13", "description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0,len(missing_cat))):
    sequence = missing_cat["description"][i]
    prediction_cat += [generate_predictions(sequence, fiction_categories)]
    isbns += missing_cat["isbn13"][i]
```

```python
print(f"Length of isbns: {len(isbns)}")
print(f"Length of prediction_cat: {len(prediction_cat)}")

missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predict_categories": prediction_cat})
```

```python
missing_predicted_df
```

```python
# merging into dataframe

books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
```
